### <font color='red'> Activity 2: Code for solving system of linear equations. </font>

### Gaussian elimination
- In LU decomposition we want to decompose original into upper and lower triangular matrices,
##### A = LU,
- A is original matrix we want to decompose
- L is lower triangular matrix (we assume it has 1-s in diagonal)
- U is upper triangular matrix

### Gaussian elimination allows three types of operations
1. Swapping two rows
2. Adding multiple of one row to other
3. Multiplying a row with a nonzero number


### Systems of linear Equations
LU Decomposition of a matrix

A set of values for x1, x2, ... xn, that satisfy all of the above equations simultaneously is said to be a solution to these equations. Case where we have n equations and n unknowns.

#### Equations in matrix-vector  Ax = b
- A is non-singular, possess an inverse
#####*Can be computationally intensive/ numerical stability.
- Use LUP-Decomposition instead!

#### LU Decomposition
- Inverse of a matrix can only be computed for a full-Rank matrix.
- x is the vector, finding whose value is the goal!
- Find two n x n matrices, L and U such,
- L is lower triangular matrix
- U is upper triangular matrix

##### Rewrite linear system,
- Define y=Ux
##### Solve following lower-triangular system first
-*Unknown y, process of Forward substitution
##### Solve upper-triangular system
- *Unknown x, process of Backward substitution

```python
def lu_decomposition(A):
  # Find number of rows
  n = A.shape[0]

  # Let L and U be new n x n matrices
  # Initialise U with 0s below the diagonal (Upper)
  U = np.zeros((n, n))

  # initialise L with 1s on the diagonal and 0s below the diagonal (Lower)
  L = np.identity(n)

  for k in range(n):
    U[k][k] = A[k][k]
    for i in range(k+1, n):
      L[i][k] = A[i][k]/ U[k][k]
      U[k][i] = A[k][i]
      for j in range(k, n):
        A[i][j] = A[i][j] - L[i][k]*U[k][j]

  return L, U


# Back Substitution
def back_substitution(U, y):
  # Find number of rows
  n = U.shape[0]

  # Vector for solution
  x = np.zeros(n)

  # Start back substitution
  for i in range(n-1,-1,-1):
    x[i] = y[i]
    for j in range(i+1,n):
      x[i] -= U[i][j]*x[j]
    x[i]/= U[i][i]

  return x

# Forward substituion
def forward_substitution(L, b):
  # Find number of rows
  n = L.shape[0]

  # Vector for solution
  y = np.zeros(n)

  # Start forward substitution
  for i in range(n):
    y[i] = b[i]
    for j in range(i):
      y[i] -= L[i][j]*y[j]

  return y

def lu_solve(A, b):
  # Perform forward substitution with L (lower)
  y = forward_substitution(L, b)

  # Perform back substitution with U (upper)
  x = back_substitution(U, y)

  return x
```

```python
# Test Case
# Define matrix A and vector b
A = np.array([[1, 2, 3], [7, 8, 9], [13, 14, 15]])
b = np.array([1, 2, 3])

# Print
print("A:")
print(A)
print()
print("B:")
print(b)
print()

# Perform LU decomposition
L, U = lu_decomposition(A)

# Print L and U matrices
print("L:")
print(L)
print()
print("U:")
print(U)
print()

# Solve
x = lu_solve(A, b)

# Print result
print("Solution:")
print(x)
```
